## Dataset Overview

> Our dataset contains data on 7.4 million trips between airports in the United States in the year 2007. It contains variables such as flight duration, distance and delays

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
data = pd.read_csv('data/2007.csv')
df = data.copy()
```

## Proportion of the air fleet in 2007

> The air fleet in 2007 is dominated by the WN company. She manages 3 trips out of 20 all over the United States. It's almost as much as the other two most popular companies, namely: AA and OO

```python
#Set visualization style
sns.set_style('whitegrid')
sns.set_context('notebook')
```

```python
order2 = df.UniqueCarrier.value_counts().index.tolist()
plt.figure(figsize =(12,8))
sns.countplot(data = df, y = 'UniqueCarrier', order = order2);
plt.xlabel('Number of Schedulled',fontsize=12)
plt.ylabel('Airlines')
plt.title('Proportion of the Airlines in 2007',fontsize=16)
plt.xlim(0, 1400000)
plt.xticks([0, 200000, 400000, 600000, 800000, 1000000, 1200000, 1400000],
           [0, '0.2M', '0.4M', '0.6M', '0.8M', '1M', '1.2M', '1.4M'])

df_counts = df.UniqueCarrier.value_counts()
for i in range(df_counts.shape[0]):
  count = df_counts[i]
  pct_string = '{:0.1f}%'.format(100*count/df.shape[0])
  plt.text(count+20, i, pct_string, va = 'center')

plt.show()
```

## Mean Distance by Airline

> American companies make trips of at least 400 miles. some are specialized on long distances, we note: CO, B6, UA and AA. Referring to the previous graph, we notice that CO and B6 are second-rate companies, not very busy, but the other two UA and AA are among those that make the most trips.

```python
order2 = df.groupby(['UniqueCarrier']).Distance.mean().sort_values(ascending=False).index
plt.figure(figsize =(12,8))
sns.barplot(data=df.sample(1000000), y='UniqueCarrier',x='Distance',order=order2)
plt.title('Mean distance by Airlines',fontsize=16)
plt.xlabel('Mean Distance (miles)',fontsize=12)
plt.ylabel('Airlines',fontsize=12)
plt.show()
```

## Mean Delay time by Airlines

> The American companies in our study in part 1 proved to be very retarding. The most famous in this sense is EV. Surely the reason why this company is little used (only 3%). It should be noted that the least solicited companies, surely new AQ and HA, are facing no major delays.

```python
order3 = df.groupby(['UniqueCarrier']).DepDelay.mean().sort_values(ascending=False).index
plt.figure(figsize =(12,8))
sns.barplot(data=df.sample(1000000), y='UniqueCarrier',x='DepDelay',order=order3)
plt.xlim(0, 25)
plt.xticks([0, 5, 10, 15, 20, 25])
plt.title('Mean delay time by Airlines',pad =5,fontsize=16)
plt.xlabel('Mean Delay Time (min)',fontsize=12)
plt.ylabel('Airlines',fontsize=12)
plt.show()
```

## Mean Deptime by Airlines

> All companies generally take off during the day. however they are all able to take off at all hours depending on schedules and delays.

```python
order4 = df.groupby(['UniqueCarrier']).DepTime.mean().sort_values(ascending=False).index
plt.figure(figsize =(12,8))
sns.boxplot(data=df.sample(1000000), x='UniqueCarrier',y='DepTime',order=order4)
plt.title('Mean depart time by Airlines',pad =5,fontsize=16)
plt.xlabel('Mean Depart Time (min)',fontsize=12)
plt.ylabel('Airlines',fontsize=12)
plt.show()
```

It specializes in medium-distance trips, with an average distance of 600 miles on all the trips it was able to make during the period. Like almost all other companies, it is subject to delays.
